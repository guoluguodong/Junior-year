```python
from PIL import Image

imgOrigin = Image.open('image/1.jpg')
width, height = imgOrigin.size
```

```python
hidden_assic = [87, 88, 16, 10, 65, 78, 6, 6, 78, 94, 12, 25, 53, 120, 49, 69, 67, 99, 48, 73, 84, 111, 39, 88, 80, 22]

hidden_binary = ''.join(format(number, '08b') for number in hidden_assic)

print(hidden_binary)
sum1 = 0
sum0 = 0
for i in range(0, len(hidden_binary)):
    if hidden_binary[i] == '1':
        sum1 += 1
    if hidden_binary[i] == '0':
        sum0 += 1
print("隐藏信息0的数目和1的数目的比例{}:{}".format(sum0 / (sum1 + sum0), sum1 / (sum1 + sum0)))
```

```python
byteArray = [0 for i in range(30)]

for y in range(height):
    for x in range(width):
        pixel = list(imgOrigin.getpixel((x, y)))
        for i in range(3):
            if 50 <= pixel[i] < 80:
                byteArray[pixel[i] - 50] += 1
print(byteArray)
ETs = [0 for i in range(15)]
for i in range(15):
    ETs[i] = sum0 / (sum1 + sum0) * byteArray[2 * i] + sum0 / (sum1 + sum0) * byteArray[2 * i + 1]
```

```python
import matplotlib.pyplot as plt

plt.bar(range(50, 80), byteArray)
plt.show()
```

```python
from PIL import Image

imgOrigin = Image.open('encrypted_image.png')
width, height = imgOrigin.size
byteArray = [0 for i in range(30)]

for y in range(height):
    for x in range(width):
        pixel = list(imgOrigin.getpixel((x, y)))
        for i in range(3):
            if 50 <= pixel[i] < 80:
                byteArray[pixel[i] - 50] += 1
print(byteArray)
print("像数值50的个数：像数值51的个数{}:{}".format(byteArray[0] / (byteArray[0] + byteArray[1]),
                                                  byteArray[1] / (byteArray[0] + byteArray[1])))
Ts = [0 for i in range(15)]
for i in range(15):
    Ts[i] = byteArray[2 * i]
import matplotlib.pyplot as plt

plt.bar(range(50, 80), byteArray)
plt.show()
```
